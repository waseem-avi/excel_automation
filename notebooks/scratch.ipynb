```python
# functions from raw.py
from datetime import datetime, timedelta
def get_payroll_window(input_date):
    date = datetime.strptime(input_date, "%d-%b-%y")
    payroll_duration = 14
    days_since_start_of_period = (date - datetime(2024, 4, 21)).days % payroll_duration
    start_date = date - timedelta(days=days_since_start_of_period)
    end_date = start_date + timedelta(days=payroll_duration - 1)
    return start_date.strftime("%m.%d"), end_date.strftime("%m.%d")

def determine_week(date_str):
    date_object = datetime.strptime(date_str, "%d-%b-%y")
    week_num = (date_object.day - 1) // 7 + 1
    return f"{date_object.strftime('%b')} {week_num}w"
```

```python
timeoff_request = pd.read_csv("../Input files/timeoffrequest_report_GT.csv")

# Normalize date format in timeoff_request for consistency
timeoff_request['Start Date'] = pd.to_datetime(timeoff_request['Start Date'], format='%b %d, %Y')
timeoff_request['End Date'] = pd.to_datetime(timeoff_request['End Date'], format='%b %d, %Y')

# Convert formatted_dates to datetime for consistency
formatted_dates = [pd.to_datetime(date, format='%d-%b-%y') for date in formatted_dates]

individual_name_dfs = {}

for name, group_df in names_group:
    # Create a copy of the DataFrame
    individual_name_df = group_df.copy()
    
    # Iterate over each date
    for date in formatted_dates:
        date_str = date.strftime('%d-%b-%y')
        # If the date is not already present in the DataFrame, add a new row
        if date_str not in individual_name_df['Date'].values:
            start_period, end_period = get_payroll_window(date_str)
            week = determine_week(date_str)
            new_row = pd.DataFrame({
                'Name': [name], 
                'Date': [date_str], 
                'Scheduled Hours': [0.0], 
                'Actual Hours': [0.0], 
                'Payroll Period': [f"{start_period} - {end_period}"], 
                'Week': [week],
                'Month':[date.strftime('%b %Y')]
            })
            individual_name_df = pd.concat([individual_name_df, new_row], ignore_index=True)
    
    # Check for time off requests and fill 'Scheduled Hours' and 'Actual Hours' with 'PTO' or 'UL'
    for i, row in individual_name_df.iterrows():
        date = pd.to_datetime(row['Date'], format='%d-%b-%y')
        request = timeoff_request[(timeoff_request['Name'] == name) &
                                  (timeoff_request['Start Date'] <= date) &
                                  (timeoff_request['End Date'] >= date) &
                                  (timeoff_request['Approved'] == "Yes")]
        
        if not request.empty:
            timeoff_type = request['Assigned Time Off Policies'].values[0]
            timeoff_value = 'UL' if timeoff_type == 'Unpaid' else 'PTO'
            individual_name_df.at[i, 'Scheduled Hours'] = timeoff_value
            #individual_name_df.at[i, 'Actual Hours'] = timeoff_value
    
    # Fill remaining missing values in 'Scheduled Hours' and 'Actual Hours' with empty strings
    individual_name_df['Scheduled Hours'] = individual_name_df['Scheduled Hours'].replace(0.0, '')
    individual_name_df['Actual Hours'] = individual_name_df['Actual Hours'].replace(0.0, '')
    
    # Sort the DataFrame by date
    individual_name_df = individual_name_df.sort_values('Date').reset_index(drop=True)
    
    # Convert 'Date' column to datetime format and sort by it
    individual_name_df['Date'] = pd.to_datetime(individual_name_df['Date'], format='%d-%b-%y')
    individual_name_df = individual_name_df.sort_values(by=['Date']).reset_index(drop=True)
    
    # Save the DataFrame for this name
    individual_name_dfs[name] = individual_name_df
```

```python
def convert_date_range(date_range_list):
    # Define a dictionary to map month numbers to month abbreviations
    month_map = {
        "01": "Jan", "02": "Feb", "03": "Mar", "04": "Apr",
        "05": "May", "06": "Jun", "07": "Jul", "08": "Aug",
        "09": "Sep", "10": "Oct", "11": "Nov", "12": "Dec"
    }

    # Function to convert a single date range
    def convert_single_date_range(date_range):
        # Split the range into start and end dates
        start_date, end_date = date_range.split(" - ")
        # Extract month and day from start and end dates
        start_month, start_day = start_date.split(".")
        end_month, end_day = end_date.split(".")
        # Map month numbers to month abbreviations
        start_month_abbr = month_map[start_month]
        end_month_abbr = month_map[end_month]
        # Return the formatted date range
        return f"{start_month_abbr} {int(start_day)} - {end_month_abbr} {int(end_day)}"

    # Apply the conversion to each date range in the list
    return [convert_single_date_range(date_range) for date_range in date_range_list]
```

```python
def calculate_total_hours_payrolls(df):
    # Ensure 'Scheduled Hours' and 'Actual Hours' columns are numeric
    df['Scheduled Hours'] = pd.to_numeric(df['Scheduled Hours'], errors='coerce').fillna(0)
    df['Actual Hours'] = pd.to_numeric(df['Actual Hours'], errors='coerce').fillna(0)
    
    # Group by 'Week' and sum the 'Scheduled Hours' and 'Actual Hours'
    total_hours = df.groupby('Payroll Period').agg({'Scheduled Hours': 'sum', 'Actual Hours': 'sum'})
    total_hours['Difference'] = total_hours['Scheduled Hours'] - total_hours['Actual Hours']
    return total_hours

# Create a dictionary to store total hours for each individual for each payroll
total_hours_by_payrolls = {}

# Calculate total hours for each individual for each payroll
for name, payroll_dfs in separate_dfs_payroll_period.items():
    total_hours_by_payrolls[name] = {}
    for payroll_period, df in payroll_dfs.items():
        total_hours_by_payrolls[name][payroll_period] = calculate_total_hours_payrolls(df)
```

```python
def calculate_total_hours_month(df):
    # Ensure 'Scheduled Hours' and 'Actual Hours' columns are numeric
    df['Scheduled Hours'] = pd.to_numeric(df['Scheduled Hours'], errors='coerce').fillna(0)
    df['Actual Hours'] = pd.to_numeric(df['Actual Hours'], errors='coerce').fillna(0)
    
    # Group by 'Month' and sum the 'Scheduled Hours' and 'Actual Hours'
    total_hours = df.groupby('Month').agg({'Scheduled Hours': 'sum', 'Actual Hours': 'sum'})
    total_hours['Difference'] = total_hours['Scheduled Hours'] - total_hours['Actual Hours']
    return total_hours

# Create a dictionary to store total hours for each individual for each payroll
total_hours_by_month = {}

# Calculate total hours for each individual for each payroll
for name, month_dfs in separate_dfs_month.items():
    total_hours_by_month[name] = {}
    for month, df in month_dfs.items():
        total_hours_by_month[name][month] = calculate_total_hours_month(df)
```

```python
def calculate_total_hours_weekly(df):
    # Ensure 'Scheduled Hours' and 'Actual Hours' columns are numeric
    df['Scheduled Hours'] = pd.to_numeric(df['Scheduled Hours'], errors='coerce').fillna(0)
    df['Actual Hours'] = pd.to_numeric(df['Actual Hours'], errors='coerce').fillna(0)
    
    # Group by 'Week' and sum the 'Scheduled Hours' and 'Actual Hours'
    total_hours = df.groupby('Week').agg({'Scheduled Hours': 'sum', 'Actual Hours': 'sum'})
    return total_hours

# Create a dictionary to store total hours for each individual for each week
total_hours_weekly = {}

# Calculate total hours for each individual for each week
for name, week_dfs in separate_dfs_week.items():
    total_hours_weekly[name] = {}
    for week, df in week_dfs.items():
        total_hours_weekly[name][week] = calculate_total_hours_weekly(df)
```
